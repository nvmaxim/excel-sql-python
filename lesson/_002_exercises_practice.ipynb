```python
nucleotides = ['A', 'T', 'G', 'C']

def get_algo(sequence):
    subsequence = []
    for element in sequence:
        new_subs = []
        for subseq in subsequence:
            new_subs.append(subseq + [element])
        subsequence.extend(new_subs)
    return subsequence
```

```python
# УПРАЖНЕНИЕ 3: "Консолидация данных (python)
# Объедините данные за Q1 и Q2, сгруппируйте по менеджерам и посчитайте общую выручку для каждого менеджера за оба периода.

import numpy as np
import pandas as pd
period_q1 = df[df["period"] == "Q1"][["manager", "revenue"]].reset_index(drop=True)

period_q2 = df[df["period"] == "Q2"][["manager", "revenue"]].reset_index(drop=True)

combined = pd.concat((period_q1, period_q2))

result = combined.groupby("manager").agg({"revenue": "sum"}).rename(columns={"revenue": "total_revenue"}).reset_index()

print(result)
```
