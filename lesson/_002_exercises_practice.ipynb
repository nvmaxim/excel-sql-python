```python
# 3.1_Задание: Сложные формулы и библиотека math
# Сторона А (Вопрос): Задача: Реализовать сложную математическую функцию f(x,y,z).
# Формула:

# Требования:
# Использовать стандартную библиотеку math.
# Помнить разницу между log (натуральный) и log10 (десятичный).

import math
def fun(x,y,z):
    num1 = 3 * (x**2)
    num2 = 4 * math.exp(x * y)
    num3 = math.sqrt(3) * math.log(z) * x * (y**4) * math.sin(y)
    denom = math.pi * math.log10(z)

    return (num1 - num2 + num3) / denom

```
