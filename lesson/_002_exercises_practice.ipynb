```python
# 3.1_Задание: Сложные формулы и библиотека math
# Сторона А (Вопрос): Задача: Реализовать сложную математическую функцию f(x,y,z).
# Формула:

# Требования:
# Использовать стандартную библиотеку math.
# Помнить разницу между log (натуральный) и log10 (десятичный).

import math
def fun(x,y,z):
    num1 = 3 * (x**2)
    num2 = 4 * math.exp(x * y)
    num3 = math.sqrt(3) * math.log(z) * x * (y**4) * math.sin(y)
    denom = math.pi * math.log10(z)

    return (num1 - num2 + num3) / denom

```

```python
# 3.2_Задание: Векторная математика (Угол между векторами)
# Сторона А (Вопрос): Задача: Найти угол между двумя n-мерными векторами в радианах.
# Указания:
# Векторы — это списки left и right.
# Использовать формулу:​
# Где a⋅b — скалярное произведение,
#  а ∣a∣ — длина (норма) вектора.

from math import sqrt, acos

def vector_angle(left, right):
    dot_product = sum(l * r for l, r in zip(left, right))
    left_norma = sqrt(sum(l**2 for l in left))
    right_norma = sqrt(sum(r**2 for r in right))
    cos_d = dot_product / (left_norma * right_norma)
    return acos(cos_d)


print(f"Тест 1 (45°): {vector_angle([1, 0, 1], [0, 0, 1]):.5f}")
print(f"Тест 2 (90°): {vector_angle([1, 0], [0, 1]):.5f}")
print(f"Тест 3 (0°):  {vector_angle([5, 5], [5, 5]):.5f}")
```
