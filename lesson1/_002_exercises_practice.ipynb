```python
a = "maabccdee"
b = "ddeffggaaz"
# Результат: "abcdefgz"

combined = a + b
unique = ""
for char in combined:
    if char not in unique:
        unique += char

sorted_result = ""
while len(unique) > 0:
    min_char = unique[0]
    for char in unique:
        if char < min_char:
            min_char = char
    sorted_result += min_char
    new_unique = ""
    removed = False
    for char in unique:
        if not removed and char == min_char:
            removed = True
            continue
        new_unique += char
    unique = new_unique
    
    
print(sorted_result)
```
