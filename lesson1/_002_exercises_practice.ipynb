```python
nucleotides = ['A', 'T', 'G', 'C', ["f"]]

def get_algo(sequence):
    subsuquence = [[]]
    for element in sequence:
        new_sub = []
        for subsueq in subsuquence:
            new_sub.append(subsueq + [element])
        subsuquence.extend(new_sub)
    return subsuquence

result = (get_algo(nucleotides))

for subseq in result:
    print(subseq)
    
```
