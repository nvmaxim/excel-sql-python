```python
# УПРАЖНЕНИЕ 5
import numpy as np

dept_report = (
    df["department"]
    .agg(
        count_manager=("manager", "count"),
        total_revenue=("revenue", "sum"),
    )
    .reset_index()
)

dept_report["revenue_share"] = (
    dept_report["total_revenue"] / dept_report["total_revenue"].sum()
)

dept_report["avg_per_manager"] = (
    dept_report["total_revenue"] / dept_report["count_manager"]
)

dept_report["high_perf"] = np.where(dept_report["avg_per_manager"] > 18000, "Yes", "No")

result = dept_report.sort_values("total_revenue", ascending=False)
print(result)
```
