```python
# УПРАЖНЕНИЕ 2
import numpy as np

result = df.copy()

result["category"] = np.where(
    df["revenue"] > 20000,
    "VIP",
    np.where(df["revenue"] > 10000, "Standart", "Basic"),
)

result = result[result["category"] == "VIP"].sort_index()

print(result)
```

```python
# УПРАЖНЕНИЕ 5
import numpy as np

result_dept = df.groupby("department").agg(
    manager_count=("manager", "count"), total_revenue=("revenue", "sum").reset_index()
)

result_dept["revenue_share"] = (
    result_dept["total_revenue"] / result_dept["total_revenue"].sum()
)

result_dept["avg_per_manager"] = (
    result_dept["total_revenue"] / result_dept["manager_count"]
)

result_dept["high_perfomance"] = np.where(
    result_dept["avg_per_manager"] > 18000, "Yes", "No"
)

result = result_dept.sort_values("total_revenue", ascending=False)

print(result)
```
