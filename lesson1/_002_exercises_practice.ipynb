```python
# Front Дано имя файла в строке filename. Как, не используя методы строк и модули (os.path.splitext и т.п.), получить расширение файла (подстроку от последней точки до конца, например для absolutely_not_a_virus.exe вывести .exe), а если точки нет или она последняя — вывести "Нет расширения"?

filename = "absolutely_not_a_virus.exe"
dod_index = -1
i = len(filename) - 1
while i > 0:
    if filename[i] == ".":
        dod_index = i
        break

if dod_index == -1 or dod_index == len(filename) - 1:
    print("no ext")
else:
    ext = ""
    i = dod_index
    while i < len(filename):
        ext += filename[i]
        i += 1
print(dod_index)
```
