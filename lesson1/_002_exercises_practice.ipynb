```python
nucleotides = ['A', 'T', 'G', 'C']

def get_algo_logic(sequence):
    subsequence = [[]]
    for element in sequence:
        new_seb = []
        for subseq in subsequence:
            new_seb.append(subseq + [element])
        subsequence.extend(new_seb)
    return subsequence

result = get_algo_logic(nucleotides)

for subseq in result:
    print(subseq)
```
