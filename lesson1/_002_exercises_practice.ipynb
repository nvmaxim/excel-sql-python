```python
# Результат: "mabcdefgz"

a = "maabccdee"
b = "ddeffggaaz"
data_input = a + b
unique = ""
for symbol in data_input:
    if symbol not in unique:
        unique += symbol
sorted_result = ""
while len(unique) > 0:
    min_symbol = unique[0]
    for symbol in unique:
        if symbol < min_symbol:
            min_symbol = symbol
    sorted_result += min_symbol
    new_unique = ""
    removed = False
    for symbol in unique:
        if not removed and symbol == min_symbol:
            removed = True
            continue
        new_unique += symbol
    unique = new_unique
print(unique, sorted_result, sep="\n")
```
