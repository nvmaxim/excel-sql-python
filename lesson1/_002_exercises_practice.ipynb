```python
def pull_turnip(characters):
    # Если список пуст, значит звать больше некого
    if not characters:
        print("...но больше никого нет! Пора тянуть!")
        return False # Репка всё еще в земле
    
    current_hero = characters[0] # Берем первого из списка
    remaining_heroes = characters[1:] # Остальные в очереди
    
    print(f"{current_hero} тянет-потянет, вытянуть не может...")
    
    # ПРОВЕРКА: Если пришла Мышка (Базовый случай)
    if current_hero == "Мышка":
        print("🐭 Мышка пришла, за хвост взялась — И ВЫТЯНУЛИ РЕПКУ!")
        return True # Ура, успех!
    
    # РЕКУРСИЯ: Зовем следующего помощника
    print(f"{current_hero} зовет следующего...")
    
    success = pull_turnip(remaining_heroes) # Функция замирает и ждет помощи
    
    if success:
        print(f"{current_hero} радуется и идет пить чай с репкой.")
        return True

# Запускаем нашу сказку!
heroes = ["Дедка", "Бабка", "Внучка", "Жучка", "Кошка", "Мышка"]
pull_turnip(heroes)
```
