```python
# Front Как, не используя методы списков и строк, по числу n построить число, в котором n записано 100 раз подряд (например, из n = 15 получить 1515...15 100 раз), а затем вывести это число и его квадрат?

n = 15
x = 0

temp = n
len_n = 0
while temp > 0:
    len_n +=1
    temp//=10


for _ in range(100):
    x = x *(10 ** len_n) + n
print(x)
print(x**2)
```
