```python
from _001_main import df_csv

df = df_csv

print(df)
```

```python
# УПРАЖНЕНИЕ 1

result = (
    df[df["department"].isin(["А", "Б"])]
    .groupby("department")
    .agg({"revenue": "mean"})
    .rename(columns={"revenue": "mean_revenue"})
    .sort_values("mean_revenue")
    .reset_index()[["department", "manager"]]
)

print(result)
```
