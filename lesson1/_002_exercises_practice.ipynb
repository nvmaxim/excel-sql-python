```python
# УПРАЖНЕНИЕ 2
import numpy as np

df["category"] = np.where(
    df["revenue"] >= 20000, "VIP", np.where(df["revenue"] >= 10000, "Standart", "Basic")
)

result = df[df["category"] == "VIP"].sort_values("revenue", ascending=False)
```

```python
# УПРАЖНЕНИЕ 3

import numpy as np
import pandas as pd

sales_q1 = df[df["period"] >= "Q1"][["manager", "revenue"]].reset_index(drop=True)
sales_q2 = df[df["period"] >= "Q2"][["manager", "revenue"]].reset_index(drop=True)

comb = pd.concat([sales_q1, sales_q2])

result = (
    comb.groupby("manager")
    .agg(total_rev=("revenue", "sum"))
    .sort_values("total_rev")
    .reset_index()
)
print(result)
```

```python
# УПРАЖНЕНИЕ 4

dept = df[(df["experience"] >= 2) & (df["department"].isin(["А", "Б"]))].copy()

dept["bonus"] = dept["revenue"] * 0.10

result = dept.rename(columns={"revenue": "sales"})
print(result)
```
