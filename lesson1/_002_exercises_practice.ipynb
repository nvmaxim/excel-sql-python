```python
nucleotides = ['A', 'T', 'G', 'C']

def generate_subs (sequence):
        subsequence = [[]]
        for element in sequence:
                new_subs = []
                for subseq in sequence:
                        new_subs.append(subseq + [element])
                subsequence.extend(new_subs)
        return subsequence

```

```python
xs = ['bob2 20', 'bob3 20', "alice 10", "bob1 20", "bertha 20", 'david 0', 'alic 10', "chris 0"]

data = []
for item in xs:
    parts = item.split()
    print(parts)
    data.append((parts[0], int(parts[1])))

data.sort()
temp = []
for key, value in data:
    temp.append((-value, key))

temp.sort()

result = [(item[1] , -item[0]) for item in temp]
print(data)
print(temp)
print(result)
```
