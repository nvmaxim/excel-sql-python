```python
# Дана строка. Проверить, все ли символы в ней уникальны (встречаются ровно 1 раз). Вывести True, если все уникальны, иначе False. set использовать нельзя.​ строка - эй,жлоб!Гдетуз?Прячьюныхсъёмщицвшкаф.


string = "эй,жлоб!Гдетуз?Прячьюныхсъёмщицвшкаф."

def get_algo(s):
    seen = []
    for symb in s:
        if symb in seen:
            return False
        seen += [symb]
    return True

print(get_algo(string))
```
