```python
# Front Дана строка. Проверить, все ли символы в ней уникальны (встречаются ровно 1 раз). Вывести True, если все уникальны, иначе False. set использовать нельзя.​ строка - эй,жлоб!Гдетуз?Прячьюныхсъёмщицвшкаф.

import sys
import io

sys.setrecursionlimit(2000)


def is_unique(text: str) -> bool:
    """
    функция по поиску уникальных символов
    """
    seen = ""
    for char in text:
        if char in seen:
            return False
        seen += char
    return True


def solve(input_data: str) -> None:
    if not input_data:
        return
    try:
        result = is_unique(input_data)
        sys.stdout.write(str(result) + "\n")
    except ValueError as e:
        sys.stdout.write(f"Errof {e}\n")


def main() -> None:
    """
    data input
    """
    try:
        raw_data = "эй,жлоб!Гдетуз?Прячьюныхсъёмщицвшкаф."
        sys.stdin = io.StringIO(raw_data)
        data = sys.stdin.read()
        solve(data)
    except EOFError:
        pass
    except Exception as e:
        sys.stdout.write(f"Error {e}\n")


if __name__ == "__main__":
    main()
```
