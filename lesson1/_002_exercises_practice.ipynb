```python
# УПРАЖНЕНИЕ 1

result = (
    df[df["department"].isin(["А", "Б"])]
    .groupby("department")
    .agg({"revenue": "mean"})
    .rename(columns={"revenje": "mean_revenue"})
    .sort_values("mean_revenue", ascending=False)
    .reset_index()[["department", "mean_revenue"]]
)

print(result)
```

```python
# УПРАЖНЕНИЕ 5
import numpy as np

dept = (
    df.groupby(["department"])
    .agg(manager_count=("manager", "count"), total_revenue=("revenue", "sum"))
    .reset_index()
)

dept["revenue_share"] = dept["total_revenue"] / dept["total_revenue"].sum()
dept["avg_per_manager"] = dept["total_revenue"] / dept["manager_count"]

dept["high_perfomance"] = np.where(dept["avg_per_manager"] >= 18000, "Yes", "No")
result = dept.sort_values("total_revenue", ascending=False)
print(result)
```
