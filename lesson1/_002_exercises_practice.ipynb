```python
a = "aabccdee"
b = "ddeffggaaz"
# Результат: "abcdefgz

combined = a + b
unique = ""
for symbol in combined:
    if symbol not in unique:
        unique += symbol

sorted_result = ""

while len(unique) > 0:
    min_sybmol = unique[0]
    for symbol in unique:
        if symbol < min_sybmol:
            min_sybmol = symbol
    sorted_result += min_sybmol
    new_unique = ""
    removed = False
    for symbol in unique:
        if not removed and symbol == min_sybmol:
            removed = True
            continue
        new_unique += symbol
    unique = new_unique

print(unique)
```
