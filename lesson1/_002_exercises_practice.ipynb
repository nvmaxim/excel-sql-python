```python
# УПРАЖНЕНИЕ 5: "Комплексный отчет" (python)
# Создайте комплексный отчет по отделам: количество менеджеров, общая выручка, доля в общей выручке, средняя выручка на менеджера и флаг высокой эффективности.
import numpy as np
result = df.groupby("department").agg(manager_count=("manager", "count"), total_revenue=("revenue", "sum")).reset_index(drop=False)

result["revenue_share"] = result["total_revenue"]  / result["total_revenue"].sum()

result["avg_per_manager"] = result["total_revenue"]  / result["manager_count"]

result["high_perfomance_flag"] = np.where(result["avg_per_manager"] >= 20000, "Yes","No")

print(result)
```
