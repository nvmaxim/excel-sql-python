```python
# Создайте комплексный отчет по отделам: количество менеджеров, общая выручка, доля в общей выручке, средняя выручка на менеджера и флаг высокой эффективности.
import pandas as pd
import numpy as np


df_report = df.groupby("department").agg(manager_count=("manager", "count"), total_revenue=("revenue", "sum")).reset_index()


df_report["revenue_share"] = df_report["total_revenue"] / df_report["total_revenue"].sum()

df_report["avg_per_manager"] =  df_report["total_revenue"] / df_report["manager_count"]

df_report["high_perfomance_flag"] = np.where(df_report["avg_per_manager"] >= 18000, "Yes", "No")

print(df_report)
```
