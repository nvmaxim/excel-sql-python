```python
from _001_main import df_csv
df = df_csv
print(df)
```

```python
# УПРАЖНЕНИЕ 1
result = (
    df[df["department"].isin(["А", "Б"])]
    .groupby("department")
    .agg({"revenue": "mean"})
    .rename(columns={"revenue": "avg_revenue"})
    .sort_values("avg_revenue", ascending=False)
    .reset_index()[["department", "avg_revenue"]]
)
print(result)
```

```python
# УПРАЖНЕНИЕ 2
import numpy as np
df["category"] = np.where(
    df["revenue"] > 20000, "VIP", np.where(df["revenue"] >= 10000, "Standart", "Basic")
)

result = df[df["category"] == "VIP"].sort_values("revenue", ascending=False)
print(result)
```
