```python
from _001_main import df_csv

df = df_csv

print(df)
```

```python
# УПРАЖНЕНИЕ 1
result = (
    df[df["department"].isin(["А", "Б"])]
    .groupby("department")
    .agg({"revenue": "mean"})
    .rename(columns={"revenue": "avg_revenue"})
    .sort_values("avg_revenue", ascending=False)
    .reset_index()[["department", "avg_revenue"]]
)
print(result)
```

```python
# УПРАЖНЕНИЕ 2
import numpy as np

df["category"] = np.where(
    df["revenue"] > 20000, "VIP", np.where(df["revenue"] >= 10000, "Standart", "Basic")
)

result = df[df["category"] == "VIP"].sort_values("revenue", ascending=False)
print(result)
```

```python
# УПРАЖНЕНИЕ 3
import pandas as pd

sales_1 = df[df["period"] == "Q1"][["manager", "revenue"]].reset_index(drop=True)
sales_2 = df[df["period"] == "Q2"][["manager", "revenue"]].reset_index(drop=True)
# print(f"таблица sales_1:\n{sales_1}\n")
# print(f"таблица sales_2:\n{sales_2}\n")
combined_sales = pd.concat([sales_1, sales_2])
print(f"таблица combined_sales:\n{combined_sales}\n")

result = (
    combined_sales.groupby("manager")
    .agg(total_revenue=("revenue", "sum"))
    .reset_index()
)
print(result)
```

```python
# УПРАЖНЕНИЕ 4
result = df[(df["experience"] >= 2) & (df["department"].isin(["А", "Б"]))].copy()
result["bonus"] = result["revenue"] * 0.10
result = result.rename(columns={"revenue": "sales"})
print(result)
```

```python
# УПРАЖНЕНИЕ 5
import numpy as np

d_re = (
    df.groupby("department")
    .agg(man_count=("manager", "count"), rev_total=("revenue", "sum"))
    .reset_index()
)

d_re["rev_share"] = d_re["rev_total"] / d_re["rev_total"].sum()
d_re["avg_per_man"] = d_re["rev_total"] / d_re["man_count"]
d_re["high_per"] = np.where(d_re["avg_per_man"] > 18000, "Да", "Нет")
d_re = d_re.sort_values("rev_total", ascending=False)
print(d_re)
```
